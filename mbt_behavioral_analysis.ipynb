## Data Visualization 

Computational replication and visualization of behavioral findings from a study investigating Mentalization-Based Therapy (MBT) effects on aggression in children with ADHD.

```python
subscales = [
    "Physical",
    "Verbal",
    "Relational",
    "Total"
]

effect_sizes = [0.33, 0.29, 0.35, 0.35]

plt.bar(subscales, effect_sizes)

plt.title("Effect Sizes of MBT on Aggression Subscales")

plt.ylabel("Partial Eta Squared")

plt.show()
```

Post-intervention comparisons between the experimental and control groups revealed clearer differences in aggression-related outcomes. Participants in the experimental group generally displayed lower aggression scores relative to the control group, suggesting that the intervention may have had a positive effect on behavioral outcomes.

```python
phases = ["pre", "post"]

experimental_values = [
    experimental_pre["physical aggression"].mean(),
    experimental_post["physical aggression"].mean()
]

control_values = [
    control_pre["physical aggression"].mean(),
    control_post["physical aggression"].mean()
]

plt.plot(phases, experimental_values, marker='o', label='experimental')
plt.plot(phases, control_values, marker='o', label='control')
plt.title("physical aggression pre/post Comparison")
plt.ylabel("Mean Aggression Score")

plt.legend()

plt.show()
```

The experimental group demonstrated a noticeable reduction in physical aggression scores following intervention, whereas the control group maintained relatively stable aggression levels across phases. This pattern may suggest that the intervention contributed to behavioral regulation improvements, particularly in aggression-related responses

## Statistical Findings And Interpretation

Descriptive comparisons revealed observable differences between the experimental and control conditions across aggression-related measures. The experimental group generally demonstrated lower post-intervention aggression scores, while the control group showed comparatively smaller changes across phases.
Visual trends suggested that participants exposed to the MBT intervention experienced reductions in physical aggression and related behavioral measures following the intervention period, However, these findings are primarily descriptive and should be interpreted cautiously without additional inferential statistical testing.
